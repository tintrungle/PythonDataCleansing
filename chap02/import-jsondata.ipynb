```python
# return the first item matches condition that less than 9
pprint.pprint(next(item for item in candidatenews if len(item) <9))
```

```python
# return the first item greater than len 9
pprint.pprint(next(item for item in candidatenews if len(item) > 9))
```

```python
# return 2 items that item length is equal 2 of number of keys
pprint.pprint([item for item in candidatenews if len(item) == 2][0:2])
```

```python
new_candidatenews = [item for item in candidatenews if len(item) > 2]
len(new_candidatenews)
```

```python
politico = [item for item in new_candidatenews if item['source'].lower() == 'politico']
len(politico)
```
