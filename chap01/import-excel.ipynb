Read Excel file

```python
import pandas as pd
```

```python
percapitaGDP = pd.read_excel('data/GDPpercapita22b.xlsx',
    sheet_name='OECD.Stat export',
    skiprows=4,
    skipfooter=1,
    usecols='A,C:W')

percapitaGDP.head()
```

```python
percapitaGDP.shape
```

```python
percapitaGDP.info()
```

```python
percapitaGDP.rename(columns={'Year':'metro'}, inplace=True)
```

```python
percapitaGDP.metro.str.startswith(' ').any()
```

```python
percapitaGDP.metro.str.endswith(' ').any()
```

```python
percapitaGDP.metro = percapitaGDP.metro.str.strip()
```

When setting errors='coerce' -> any elements cannot be converted to numbers will be replaced with 'NaN' (Not a number)

```python
for col in percapitaGDP.columns[1:]:
    percapitaGDP[col] = pd.to_numeric(percapitaGDP[col], errors='coerce')
    percapitaGDP.rename(columns={col: 'pcGDP'+col}, inplace=True)
```

```python
percapitaGDP.head()
```

```python
percapitaGDP.dtypes
```

```python
percapitaGDP.describe()
```

Setting how='all' to choose rows only with all columns that are missing data

```python
percapitaGDP.dropna(subset=percapitaGDP.columns[1:], how='all', inplace=True)
```
