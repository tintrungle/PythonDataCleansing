* Uses the statistical rule that in a normal distribution, about 99.7% of data falls within 3 standard deviations of the mean
* Any value more than 3 standard deviations above the mean is considered an outlier

```python
highwages = nls97.loc[nls97.wageincome20 > nls97.wageincome20.mean() + (nls97.wageincome20.std() * 3), ['wageincome20']]
highwages
```
